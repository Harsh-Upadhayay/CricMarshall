```python
import requests

# set the match ID for the commentary you want to retrieve
match_id = 1269635

# make request to API
response = requests.get(f'https://site.web.api.espn.com/apis/site/v2/sports/cricket/80331/playbyplay?contentorigin=espn&event={match_id}&pageview=full&lang=en&region=in')

# extract commentary data from response
data = response.json()
commentary_items = data.get('commentary', {}).get('items', [])

# iterate over each item in the commentary and print the text
for item in commentary_items:
    print(item.get('description', ''))
```

```python
import requests
import json
```

```python
url = 'https://site.web.api.espn.com/apis/site/v2/sports/cricket/8039/playbyplay'
params = {'contentorigin': 'espn', 'limit': '1'}
```

```python

# make API request
response = requests.get(url, params=params)
```

```python
# extract commentary data from response
data = response.json()
commentary_link = data['commentary']['items'][0]['links'][0]['href']
```

```python
# make API request to commentary link
response = requests.get(commentary_link)
data = response.text
commentary = json.loads(data)['commentary']

# print the commentary
print(commentary)
```
