```python
# Define the CSV file path
csv_file_path = "IPL_Match_Highlights_Commentary.csv"
```

```python
# Define a function to generate summaries
def generate_summary(overs, comments):
    # Set up the input prompt
    input_prompt = prompt.format(overs=overs, comments=comments)

    # Generate the summary using GPT-3
    response = openai.Completion.create(
        engine="davinci",
        prompt=input_prompt,
        temperature=0.5,
        max_tokens=100,
        n=1,
        stop=None,
        timeout=60,
    )
    # Extract the summary from the API response
    summary = response.choices[0].text.strip()

    return summary
```

```python
# Read the CSV file and generate summaries for each match
summaries = defaultdict(str)
with open(csv_file_path, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        match_id = row['Match_id']
        overs = row['over_num']
        comment = row['comment']
        summaries[match_id] += comment + ' '
```

```python
# Generate summaries for each match
for match_id, summary_text in summaries.items():
    summary = generate_summary(summary_text.count(' '), summary_text)
    print(f"Match {match_id} summary: {summary}\n")
```
