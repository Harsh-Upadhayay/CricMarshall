```python
# Load the pre-trained T5 model and tokenizer
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load the ball-by-ball commentary data from CSV file
import io
df = pd.read_csv(io.BytesIO(uploaded['IPL_Match_Highlights_Commentary.csv']))
```

# New Section

```python
# Group the commentary by match and concatenate into a single string for each match
match_text = df.groupby(['Match_id', 'Team'])['Commentary'].apply(lambda x: ' '.join(x)).tolist()
```

```python
for i, text in enumerate(match_text):
    input_ids = tokenizer.encode('summarize: ' + text, return_tensors='pt', max_length=512, truncation=True)
    summary_ids = model.generate(input_ids=input_ids, max_length=100, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    match_id, team = df['Match_id'].unique()[i], df['Team'].unique()[i]
    print(f"summary: {summary}")
```
