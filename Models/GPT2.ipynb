```python
import pandas as pd
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
# Load the data
data = pd.read_csv('Cricket_Commentary.csv')
```

```python
# Initialize the GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Define the device to run the model on
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Define the maximum length of the summary to generate
max_summary_length = tokenizer.model_max_length - 2  # Account for special tokens
```

```python
# Define the function to generate summaries
def generate_summary(comment):
    # Encode the input comment using the tokenizer
    input_ids = tokenizer.encode(comment, return_tensors='pt').to(device)

    # Generate the summary using the model
    output_ids = model.generate(input_ids, max_length=max_summary_length, pad_token_id=tokenizer.eos_token_id)
    output_str = tokenizer.decode(output_ids.squeeze(), skip_special_tokens=True)

    return output_str
```

```python
# Loop through each match in the dataset
for match_id, match_data in data.groupby('Match_id'):
    print('Match ID:', match_id)
    print('Match Summary:', '-'*50)
    
    # Initialize the summary for the match
    match_summary = ''
    
    # Loop through each over in the match
    for over_num, over_data in match_data.groupby('over_num'):
        # Combine the comments in the over into a single string
        comment_str = ' '.join(over_data['comment'].tolist())

        # Generate the summary for the over
        summary_str = generate_summary(comment_str)

        # Append the over summary to the match summary
        match_summary += summary_str
        print(match_summary)
        
    # Print the match summary
    print(match_summary)
    print('\n\n')
```
